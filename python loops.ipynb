**Given a list, print even numbers in list**

```python
list1=[2,3,6,8,12,13,17]
for num in list1:
  if num%2==0:
    print(num)
```

```python
list1=[2,3,6,8,12,13,17]
list2=[]
for num in list1:
  if num%2==0:
    list2.append(num)
print(list2)
```

**Factorial of a given number by (while loop)**

```python
num= int(input("enter the number:"))
factorial=1
while num>0 :
  factorial *= num
  num-=1

print(factorial)
```

**Factorial of a given number by (for loop)**

```python
num= int(input("enter the number:"))
factorial=1

for i in range(1,num+1):
  factorial= factorial * i

print(factorial)
```

**Reversing a list using two pointer technique**

```python
list2=[2,5,8,12,17]
low=0
high= len(list2) - 1
while low < high :
 ## swaping the low and high pointers
  list2[low], list2[high] = list2[high], list2[low]

 ##incrementing the low pointer and decrementing the high pointer
  low+=1
  high-=1

print(list2)


```
