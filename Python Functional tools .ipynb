**Normal function declaration**

```python
def cube(x):
  return x*x*x

print(cube(3))
```

**Lambda function**

```python
lambda_cube= lambda x:x*x*x

lambda_cube(3)
```

**addition using lambda function**

```python
lambda_sum = lambda x,y:x+y
lambda_sum(2,3)
```

**Sort a list of dictionary on the basis of specific key**

```python
employee_records = [{"name":"John", "age":29 , "occupation":"engineer" },
                    {"name":"mohan", "age":25 , "occupation":"doctor" },
                    {"name":"ram", "age":25 , "occupation":"scientist" },
                    {"name":"hari", "age":25 , "occupation":"magician" },
                    {"name":"alex", "age":23 , "occupation":"engineer" },]
```

**data sorted on the basis of key 'age'**

```python
sorted(employee_records, key=lambda x:x['age'])
```

**sort the data as key value as age,name**

```python
sorted(employee_records, key=lambda x:(x['age'] , x['name']))
```

**Implement a function which result in square of every number in the avalaible list**

```python
l1=[2,4,5,7,9]
new_l1=[]

for i in l1:
  new_l1.append(i**2)

print(new_l1)
```
