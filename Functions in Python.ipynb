```python
def reverse(list1):
  low,high = 0,len(list1)-1
  while low<high :
    list1[low],list1[high] = list1[high],list1[low]

    low+=1
    high-=1
  return list1



list1=[2,5,7,12,10]
print(reverse(list1))
```

**fibonacci series**

```python
def febonacci_function(n):
  result=[]
  a,b=0,1
  for i in range(n):
    result.append(a)
    a,b= b,a+b
  return result

print(febonacci_function(5))



```

Letâ€™s trace what happens: in the above code

Start: a = 0, b = 1, result = []

Iteration 1: result = [0], a = 1, b = 1

Iteration 2: result = [0, 1], a = 1, b = 2

Iteration 3: result = [0, 1, 1], a = 2, b = 3

Iteration 4: result = [0, 1, 1, 2], a = 3, b = 5

Iteration 5: result = [0, 1, 1, 2, 3], a = 5, b = 8
