**factorial of the given no**

```python
def fact(n):
  ## base condition
  if n<=1:
    return 1
  ## recursive call
  else:
    return n * fact(n-1)

## driver code
n = int(input("Enter a number: "))
print(fact(n))
```

**febonacci series**

```python
def feb(n):
  ##base condition
  if n<=1:
    return n
  ##recursive call
  else:
    return feb(n-1) + feb(n-2)

##driver code
n = int(input("Enter a number: "))
print(feb(n))
```

**Sum of digits in the number**

```python
def sum(n):
  ## base condition
  if n<10:
    return n
  ##recursive call
  else:
    return (n%10) + sum(n//10)

## driver code
n = int(input("Enter a number: "))
print(sum(n))
```
