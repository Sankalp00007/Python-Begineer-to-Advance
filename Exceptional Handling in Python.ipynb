**Handling the error of above code via using try , except statements**

```python
try :
  a=int(input())
  b=int(input())
  c=a/b
  print(c)
except:
  pass
print("Hello World")
```

**specific to the exception**

```python
try :
  a=int(input())
  b=int(input())
  c=a/b
  print(c)
except Exception as e:
  print(e)
print("Hello World")
```

**index error**

```python
try:
  l=[10,20,40,90,50]
  l[10]
except Exception as e:
  print(e)
```

**Key Value Error**

```python
try:
  records={"name" : "sankalp" , "email":"sankalp@gmail.com" }
  records["age"]

except Exception as e:
  print(e)
```

**else block and finally block**

```python
try :
  a=int(input())
  b=int(input())
  c=a/b
  print(c)
except Exception as e:
  print(e)

else:
  print("it will execute only when try block runs successfully")

finally:
  print("this block will execute always")
```

**Real time example of data-preprocessing**

```python
try:
  file_path="/content/demo.txt"
  f=open(file_path,"r")
  content=f.read()
  print(content)
except Exception as e:
  print(e)
```
