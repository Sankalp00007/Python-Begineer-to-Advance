## Approach 1: where we put the condition m1== m2 (slopes of the points are equal if they are collinear)

```python
def check_collinear(x1,y1,x2,y2,x3,y3):
    if (y2-y1)*(x3-x2) == (y3-y2)*(x2-x1):
        print("the given points are collinear")
    else:
        print("not")

# driver code
x1,y1,x2,y2,x3,y3 = 1,6,1,0,1,9
print(check_collinear(x1,y1,x2,y2,x3,y3))
```

## Approach 2 : if the area of triangle ==0 then the points the collinear else not

```python
def check_collinear(x1,y1,x2,y2,x3,y3):
    area = 0.5* ( x1*(y2-y3) + x2*(y3-y1) + x3*(y1-y2) )
    if area == 0:
        print("the given points are collinear")
    else:
        print("not")

# driver code
x1,y1,x2,y2,x3,y3 = 1,6,1,0,1,9
print(check_collinear(x1,y1,x2,y2,x3,y3))
```
