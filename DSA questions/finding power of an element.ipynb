```python
def findPowerofElement(a,n):
    #small problem
    if n==1:
        return a
    elif n== 0:
        return 1
    else:
        #big problem
        #1.divide
        mid= n//2
        #2.Conquer
        b = findPowerofElement(a,mid)
        # multiply two times i.e same for right side
        result = b*b
        if n%2==0:
            return result
        ## for odd num    
        return result * a 
        
#driver code 
a=2
n=17
# function calling
result = findPowerofElement(a,n)
print("the power of an element is:",result)
```
