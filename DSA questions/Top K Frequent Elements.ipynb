```python
from collections import Counter
import heapq

def topKfrequentelements(arr, k):
    # If k equals the total number of unique elements, return all unique elements
    if k == len(arr):
        return list(set(arr))

    count = Counter(arr)
    print(count)  # Optional: to show frequency counts

    # Get k elements with highest frequency
    return heapq.nlargest(k, count.keys(), key=count.get)

# Driver Code
arr = [1, 1, 1, 1, 2, 2, 2, 3]
k = 2

# Function calling
result = topKfrequentelements(arr, k)
print("The top k frequent elements are:", result)
```
