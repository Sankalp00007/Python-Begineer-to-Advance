```python
## Method definition of max and min function
def findMaxandMin(arr,i,j):
    #small problem
    if i==j:
        max_val = arr[i]
        min_val = arr[i]
    #two element condition
    elif i== j-1:
        if arr[i] < arr[j]:
            min_val = arr[i]
            max_val = arr[j]
        else:
            max_val = arr[i]
            min_val = arr[j]
    # large problem i.e n > 2
    else:
        # divide and conquer approach
        #1.divide
        mid= i+ (j-i)//2
        #2.recursion
        max1,min1 = findMaxandMin(arr,i,mid)
        max2,min2 = findMaxandMin(arr,mid+1,j)
        #3. Combining
        # for max
        if max1 < max2:
            max_val = max2
        else:
            max_val = max1
        #for min    
        if min1< min2:
            min_val = min1
        else:
            min_val = min2
    return max_val,min_val  

## driver code
arr = [20,39,45,65,21,44,89,92]
i=0
j= len(arr) - 1 
## function calling
max_val,min_val = findMaxandMin(arr,i,j)
print("Maximum and Minimum value of the array is:",max_val,min_val)
```
