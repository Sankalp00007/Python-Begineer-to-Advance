```python
## time complexity: o(n)
def two_pointer(arr,target):
    l=0
    r= len(arr)-1
    while(l<=r):
        if arr[l] + arr[r]== target:
            return (l,r)
        elif arr[l] + arr[r] > target:
            r= r-1
        else:
            l= l+1
    return -1,-1        
            
## driver code
arr=[20,40,60,80,90,120,240]
target = 210
result = two_pointer(arr,target)
print(result)
```
