```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset('tips')
```

```python
data.head()
```

```python
data.shape
```

**Visualizing the relationshipp between  tip and total_bill value using relplot(relationship plot)**

```python
sns.relplot(x='tip',y='total_bill',data=data)
```

```python
sns.relplot(x='tip',y='total_bill',data=data,hue='day')
```

**Categorization plot (catplot)**

```python
sns.catplot(x='day',y='tip',data=data)
```

```python
sns.catplot(x='smoker',y='tip',data=data,)
```

```python
sns.catplot(x='day',y='tip',data=data,hue='sex',kind='box')
```

**distribution plot**

```python
sns.distplot(data.tip)
```
